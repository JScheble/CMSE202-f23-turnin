&#9989; **Question 2.2 (3 points)** Now, construct a graph using Networkx. Display the graph and use the position information to place the water stations at the correct locations. 


````python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Load adjacency matrix usingb Pandas
adj_df = pd.read_csv('h20_adj.txt', header=None)
adj_matrix = adj_df.values

# Load node positions using Pandas
positions_df = pd.read_csv('h20_position.txt', header=None)
positions = positions_df.values

# Create a graph from the adjacency matrix
G = nx.from_numpy_matrix(adj_matrix)

# Set the positions of nodes using the loaded coordinates
pos = {i: (positions[i, 0], positions[i, 1]) for i in range(len(positions))}

# Draw the graph with node positions
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=500, node_color='skyblue', font_color='black', font_size=8)

# Display the graph
plt.show()
````

&#9989; **Question 2.3 (2 points)** Print the total number of nodes in the network and the number of edges


````python
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()

print(f"Total number of nodes: {num_nodes}")
print(f"Total number of edges: {num_edges}")
````

&#9989; **Question 2.4 (2 points)** In the interest of having an efficient water delivery system, we would like to know the shortest path (in terms of fewest number of edges) from one water station to another station.  Networkx has a convenient method to find these paths.  Using this method, answer the following questions:  
* Taking the shortest path, how many nodes are in-between node 27 and 20?  
* Likewise, taking the shortest path, how many edges (water pipes) are in-between node 27 and 20?


````python
# Find the shortest path from node 27 to node 20
shortest_path_nodes = nx.shortest_path(G, source=27, target=20)

# Calculate the number of nodes and edges in the shortest path
num_nodes_in_path = len(shortest_path_nodes)
num_edges_in_path = num_nodes_in_path - 1  # Since each edge connects two nodes

# Print the results
print(f"Number of nodes in the shortest path from node 27 to 20: {num_nodes_in_path}")
print(f"Number of edges in the shortest path from node 27 to 20: {num_edges_in_path}")
````

&#9989; **Question 2.5 (1 point)** Taking this shortest path between node 27 and 20, what is the physical distance between these water stations?  In other words, how long is the pipe that runs along this shortest path?  

Hint: Use the information in the position data file to determine the physical (euclidean) distance between nodes.

Note: If you are stumped on this, recommend completing other sections of exam before working on this (only worth 1 point).  


````python

# Create a graph from the adjacency matrix
G = nx.from_numpy_matrix(adj_matrix)

# Find the shortest path from node 27 to node 20
shortest_path_nodes = nx.shortest_path(G, source=27, target=20)

# Calculate the physical distance along the shortest path
physical_distance = 0
for i in range(len(shortest_path_nodes) - 1):
    node1 = shortest_path_nodes[i]
    node2 = shortest_path_nodes[i + 1]
    coord1 = positions[node1]
    coord2 = positions[node2]
    distance = np.linalg.norm(coord2 - coord1)  # Euclidean distance
    physical_distance += distance

# Print the result
print(f"Physical distance along the shortest path from node 27 to 20: {physical_distance}")
````
